```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Pl_Data.xlsx')
```

```python
data.head()
```

```python
data.shape
```

adjusting data

```python
data["now_cost"]=data["now_cost"]/10
data['goals_involvements']=data['goals_scored']+data['assists']
data['expected_goals_involvements']=data['expected_goals']+data['expected_assists']
```

```python
data
```

Players with an average 30 minutes per game (30*38 matchs per season)

```python
filtered_players = data[data['minutes'] > 1140]

# Calculate the average of 'goals_scored' for these players
average_goals_scored = filtered_players['goals_scored'].mean()
average_goals_scored
```

```python
filtered_players.head()
```

```python
high_scorers = data[data['goals_scored'] > average_goals_scored]
```

```python
high_scorers
```

Number of players who scored more than the average: 100

```python
forwards = high_scorers[high_scorers['position'] == 'FWD']

```

```python
forwards.head()
```

```python
# Calculate the percentage of forwards among all high scorers
percentage_forwards = (len(forwards) / len(high_scorers)) * 100

print(f"Percentage of forwards among high scorers: {percentage_forwards:.2f}%")
```

```python
grouped = forwards.groupby(['goals_scored', 'now_cost']).size().reset_index(name='count')


dot_size = grouped['count'] * 50 

# Create a scatter plot with scaled dot size
plt.figure(figsize=(10, 6))
plt.scatter( grouped['now_cost'],grouped['goals_scored'], s=dot_size, color='blue', alpha=0.7)
plt.title('Goals Scored vs. Now Cost')
plt.ylabel('Goals Scored')
plt.xlabel('Now Cost')
plt.title('Goals Scored vs. Cost for Forward Players')
plt.text(5.5, 35, 'Conclusion: As goals scored increase, the cost tends to increase', fontsize=12, color='red')
```

```python
midfielders = high_scorers[high_scorers['position'] == 'MID']

# Calculate the percentage of forwards among all high scorers
percentage_midfielders = (len(midfielders) / len(high_scorers)) * 100

```

```python
midfielders.head()
```

```python
print(f"Percentage of forwards among high scorers: {percentage_midfielders:.2f}%")
```

```python
grouped = midfielders.groupby(['goals_scored', 'now_cost']).size().reset_index(name='count')


dot_size = grouped['count'] * 50 

# Create a scatter plot with scaled dot size
plt.figure(figsize=(10, 6))
plt.scatter(grouped['now_cost'],grouped['goals_scored'], s=dot_size, color='blue', alpha=0.7)
plt.title('Goals Scored vs. Now Cost')
plt.ylabel('Goals Scored')
plt.xlabel('Now Cost')
plt.title('Goals Scored vs. Cost for Midfielders Players')
plt.text(5, 18, 'Conclusion: As goals scored increase, the cost tends to increase', fontsize=12, color='red')
```

```python
defenders = high_scorers[high_scorers['position'] == 'DEF']

# Calculate the percentage of forwards among all high scorers
percentage_defenders = (len(defenders) / len(high_scorers)) * 100



print(f"Percentage of defenders among high scorers: {percentage_defenders:.2f}%")
```

```python
more_than_expected = filtered_players[filtered_players['goals_involvements'] > filtered_players['expected_goals']+filtered_players['expected_assists']]
```

```python
more_than_expected[['name','goals_involvements','expected_goals_involvements','now_cost','position']]
```

143 players have goals involements more than the expected

```python
grouped = more_than_expected.groupby(['goals_involvements', 'now_cost']).size().reset_index(name='count')


dot_size = grouped['count'] * 50 

# Create a scatter plot with scaled dot size
plt.figure(figsize=(10, 6))
plt.scatter(grouped['now_cost'],grouped['goals_involvements'], s=dot_size, color='blue', alpha=0.7)
plt.title('Goals involvements vs. Now Cost')
plt.ylabel('Goals involvements')
plt.xlabel('Now Cost')
plt.title('Goals involvements vs. Cost for players have goals involvements more than the expected')
plt.text(5, 42, 'Conclusion: As goals involvements increase, the cost tends to increase', fontsize=12, color='red')
```

```python
more_than_expected_forwards=more_than_expected[more_than_expected['position']=='FWD']
more_than_expected_midfielders=more_than_expected[more_than_expected['position']=='MID']
more_than_expected_defenders=more_than_expected[more_than_expected['position']=='DEF']
```

```python
fwd_percentage=len(more_than_expected_forwards) / len(more_than_expected) * 100
mid_percentage=len(more_than_expected_midfielders) / len(more_than_expected) * 100
def_percentage=len(more_than_expected_defenders) / len(more_than_expected) * 100
```

```python
plt.pie([fwd_percentage,mid_percentage,def_percentage], labels=['Forwards','Midfielders','Defenders'], autopct='%1.1f%%', startangle=140)
plt.title('goals involvements more than expected distibution')
```
