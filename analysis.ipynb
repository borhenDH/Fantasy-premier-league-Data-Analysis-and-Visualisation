Players with an average 30 minutes per game

```python
filtered_players = data[data['minutes'] > 1140]

# Calculate the average of 'goals_scored' for these players
average_goals_scored = filtered_players['goals_scored'].mean()
average_goals_scored
```

```python
high_scorers = data[data['goals_scored'] > average_goals_scored]
print(high_scorers)

count_high_scorers = len(high_scorers)
print(f"Number of players who scored more than the average: {count_high_scorers}")
```
