```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Pl_Data.xlsx')
```

```python
data.head()
```

An Abundance of the Dataset

```python
data.shape
```

The dataset consists of 778 rows and 75 columns.

adjusting data

```python
data["now_cost"]=data["now_cost"]/10
data['goals_involvements']=data['goals_scored']+data['assists']
data['expected_goals_involvements']=data['expected_goals']+data['expected_assists']
```

```python
data
```

the dataset after the adjustments

```python
data.shape
```

```python
filtered_players = data[data['minutes'] > 1140]

filtered_players.head()
```

Players with an average 30 minutes per game (30*38 matchs per season)

```python

average_goals_scored = filtered_players['goals_scored'].mean()
average_goals_scored
```

Average Goals of Players with 30 Minutes or More Per Game

```python
high_scorers = data[data['goals_scored'] > average_goals_scored]
```

```python
high_scorers
```

Players Who Scored Above the Average

```python
forwards = high_scorers[high_scorers['position'] == 'FWD']
```

```python
forwards.head()
```

Forwards who scored above the average

```python
grouped = forwards.groupby(['goals_scored', 'now_cost']).size().reset_index(name='count')


dot_size = grouped['count'] * 50 

# Create a scatter plot with scaled dot size
plt.figure(figsize=(10, 6))
plt.scatter( grouped['now_cost'],grouped['goals_scored'], s=dot_size, color='blue', alpha=0.7)
plt.title('Goals Scored vs. Now Cost')
plt.ylabel('Goals Scored')
plt.xlabel('Now Cost')
plt.title('Goals Scored vs. Cost for Forward Players')
plt.text(5.5, 35, 'Conclusion: As goals scored increase, the cost tends to increase', fontsize=12, color='red')
```

```python
midfielders = high_scorers[high_scorers['position'] == 'MID']
```

```python
midfielders.head()
```

```python
grouped = midfielders.groupby(['goals_scored', 'now_cost']).size().reset_index(name='count')


dot_size = grouped['count'] * 50 

# Create a scatter plot with scaled dot size
plt.figure(figsize=(10, 6))
plt.scatter(grouped['now_cost'],grouped['goals_scored'], s=dot_size, color='blue', alpha=0.7)
plt.title('Goals Scored vs. Now Cost')
plt.ylabel('Goals Scored')
plt.xlabel('Now Cost')
plt.title('Goals Scored vs. Cost for Midfielders Players')
plt.text(5, 18, 'Conclusion: As goals scored increase, the cost tends to increase', fontsize=12, color='red')
```

Midfielders who scored above the average

```python
defenders = high_scorers[high_scorers['position'] == 'DEF']
```

```python
percentage_defenders = (len(defenders) / len(high_scorers)) * 100

percentage_midfielders = (len(midfielders) / len(high_scorers)) * 100

percentage_forwards = (len(forwards) / len(high_scorers)) * 100
colors = ['red', 'blue', 'purple']
plt.pie([percentage_forwards,percentage_midfielders,percentage_defenders], labels=['Forwards','Midfielders','Defenders'],colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Percentage of Goals by position')
```

```python
defenders_stats = pd.DataFrame()

defenders_stats[[ 'name','points','cost','clean_sheets']] = data[data['position'] == 'DEF'][[ 'name','total_points','now_cost','clean_sheets']]
defenders_stats = defenders_stats.sort_values(by='points', ascending=False)
```

```python
defenders_stats
```

```python
defenders_stats = defenders_stats[defenders_stats['points'] != 0]
```

```python
defenders_stats
```

```python
grouped = defenders_stats.groupby(['points', 'cost']).size().reset_index(name='count')


dot_size = grouped['count'] * 50 

# Create a scatter plot with scaled dot size
plt.figure(figsize=(10, 6))
plt.scatter(grouped['cost'],grouped['points'], s=dot_size, color='blue', alpha=0.7)
plt.title('Total points vs. Cost')
plt.ylabel('Total points')
plt.xlabel('Cost')
plt.title('Total points vs. Cost for players have goals involvements more than the expected')
plt.text(4.5, 175, 'Conclusion: As points increase, the cost tends to increase', fontsize=12, color='red')
```

```python
correlation = defenders_stats['points'].corr(defenders_stats['clean_sheets'])
print(correlation)
```

```python
correlation = defenders_stats['points'].corr(defenders_stats['clean_sheets'])


plt.figure(figsize=(8, 6))
plt.scatter(defenders_stats['clean_sheets'], defenders_stats['points'], c='b', label=f'Correlation: {correlation:.2f}')
plt.title('Correlation between Points and Clean Sheets for Defenders')
plt.xlabel('Clean Sheets')
plt.ylabel('Points')
plt.legend()
plt.grid(True)


plt.show()
```

The scatter plot illustrates a very strong positive correlation (correlation coefficient ≈ 0.95) between 'points' and 'clean_sheets' for defenders in the 'defenders_stats' DataFrame. This indicates that as clean sheets increase, defender points also rise significantly, making clean sheets a crucial factor for maximizing Fantasy Premier League points.

The correlation coefficient of approximately -0.947 suggests a strong negative correlation between the 'points' and 'standings' of the teams. In other words, as the 'points' increase, the 'standings' tend to decrease, indicating that teams with higher points typically have better standings
